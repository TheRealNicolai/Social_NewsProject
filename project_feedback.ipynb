## Research question
To what extent do news articles form distinct communities based on semantic similarity, and how are the resulting community structures, sentiments, and emotional profiles related to the 1) Country of Origin of the news articles and 2) Socio-Demographic characteristics (age, language, income group, and average years of schooling) of the populations within those identified countries?

### Sub-questions
>- Community Structure: How many distinct communities emerge in the news article network, and what are their defining characteristics (e.g., density, size, core articles)?

>- Country/Topic Influence: Is there a statistically significant correlation between community membership and the country of origin or the primary topic of the articles?

>- Sentiment and Emotion: Do the identified news communities exhibit measurable differences in overall sentiment (positive, negative, neutral) and specific emotional content (e.g., anger, fear, joy)?

>- Demographic Relationship: For communities strongly associated with specific countries, what is the relationship between their collective sentiment/emotional profile and the socio-demographic indicators (age, language, income, schooling) of those countries?
