## World data
Our geo data set i downloaded : https://naciscdn.org/naturalearth/110m/cultural/ne_110m_admin_0_countries.zip

### Names for country:
- **ADMIN**
- - Administrative name of the country in "short-form english"
- **NAME**
- - Cartographic name. Often times same as ADMIN. (e.g. *'Dem. Rep. Congo'*)
- **NAME_LONG**
- - Long form of country name (e.g. *'Democratic Republic of the Congo'*)
- **ABBREV**
- - Abbreviation of name. E.g. *'U.S.A'*
- **FORMAL_EN / FORMAL_FR**
- - Long and formal titel in english and french. E.g. *'Kingdom of Denmark'*, *'République française'*
- **NAME_CIAWF**
- - Country name in CIA World Factbook
- **NAME_SORT**
- - Name optimized for sorting
- **NAME_ALT**
- - Alternative names
- **NAME_XX**
- - Name in other language. E.g. *'NAME_DE'* for German name for a country.

### Population, economic and other useful data
- **POP_EST**
- - Estimated population
- **POP_RANK**
- - Ranking of population
- **POP_YEAR**
- - Year the estimated population is taken
- **GDP_MD**
- - Gross Domestic Product in millions of dollars (MD)
- **GDP_YEAR**
- - Year the Gross Domestic Product is taken.
- **ECONOMY**
- - E.g. *'1. Developed region: G7'*, *'7. Least developed region'* etc.
- **INCOME_GRP**
- - One of these groups: *'1. High income: OECD', '2. High income: nonOECD', '3. Upper middle income', '4. Lower middle income', '5. Low income'*
- **WIKIDATAID**
- - Wikidata-ID (e.g. Q191 for Estland, Q954 for Zimbabwe). Could be used if information from wiki is wanted.


### Political information
- **SOVEREIGNT**
- - Sovereign entity. E.g. for *'Greenland'* it is *'Denmark'*
- **SUBUNIT**
- - E.g. *'Scotland'* as subunit of *'United Kingdom'*
- **TYPE**
- - E.g. *'Sovereign country'*, *'Disputed'*, *'Dependency'* etc.
- **BRK_NAME**
- - What the breakaway entity calls it self
- **BRK_GROUP**
- - Group of breakaway units (almost never used)
- **NOTE_BRK**
- - Notes about breakaway status. E.g. *'Self admin.; Claimed by China'*
- **FCLASS_xx**
- - xx is a big entity. Whether the entity recognizes the country. e.g. *'Admin-0 breakaway and disputed', 'Unrecognized', 'Admin-0 country'* 

### Geopgraphy
- **CONTINENT**
- - The continent where the country is place
- **HOMEPART**
- - $1$ is the polygon is part of the homeland. Other value if not (e.g. island across the sea)
- **REGION_UN**
- - UN's assigned region for the country
- **SUBREGION**
- - E.g. *Northern Europe, Eastern Africa*
- **REGION_WB**
- - Region name according to the world bank
- **TINY**
- - Flagging tiny countries and microstates

### Mapping details
- **geometry**
- - The geometry of the country. Either in Multipolygon or Polygon.
- **scalerank**
- - Number that tells us how "important" a country is on generalised world maps. Smaller is more important. *'French Southern and Antarctic Lands'* is $3$. the rest is $1$.
- **LABELRANK**
- - Kind of like scalerank but specified to *labels*, so low values means that the label will be shown on lower zoom levels.
- **GEOU_DIF, GEOUNIT, GU_A3**
- - 'Geographic unit'. For places like islands where geography and admin might not be 1:1
- **MAPCOLOR7**
- - Gives all countries one of seven colors so no neighboring countries gets the same color. E.g. for political world maps.
- **MAPCOLOR8, MAPCOLOR9, MAPCOLOR13**
- - Same but with different number of colors.
- **MIN_ZOOM, MIN_LABEL, MAX_LABEL**
- - At what zoom level the country should be drawn
- **LABEL_X, LABEL_Y**
- - Coordinates for labels

### Codes and administrative categories for countries:
- **SU_DIF and SU_A3**
- - Whether country is subunit of other country, and 3-letter code for subunit.
- **ADM0_A3**
- - Natural Eart's own 3-letter code for each country (often same as ISO-3)
- **ISO_A3 / ISO_A2 / ISO_N3**
- - ISO codes like: A2 for Denmark is *DK*, A3 is *DNK* and N3 is $208$
- **SOV_A3**
- - $3$ letter code for sovereign entity.
- **ADM0_DIF**
- - $0$ if *SOVEREIGNT* and *ADMIN* is the same, $1$ if they are different
- **featurecla**
- - Feature class. They are all just *'Admin-0 country'*
- **LEVEL**
- - Administrative level ($2$ for almost all. Only *'Kazakhstan'* and *'Cuba'* are level 1) 
- **TLC**
- - Some internal Natural-Earth data field. Not really clear.
- **BRK_DIFF**
- - If there is differences between 'breakaway' names/code and standard name
- **BRK_A3** 
- - $3$-letter code for the breakaway unit (if any)
- **POSTAL**
- - Postal-code like codes for countries
- **NOTE_ADM0**
- - Notes about admin-0 status
- **FIPS_10**
- - Old system for country codes
- **ISO_A2 / ISO_A2_EH / ISO_A3 / ISO_A3_EH / ISO_N3 / ISO_N3_EH**
- - Different ISO codes for the countries. EH means *'empirical hierarchy'*
- **UN_A3**
- - UN's 3-letter country code
- **WB_A2 / WB_A3**
- - Country code from the world bank
- **ADM0_ISO**
- - ISO code for the admin-0 unit
- **ADM0_A3_xx**
- - The code a big entity uses for the country. xx is the big entitys like US, FR, RU, UN etc. If an entity doesn't recognize the country the code is $-99$
- **NE_ID**
- - Internal Natural Earth ID
